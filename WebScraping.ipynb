### Scrapping the data from beautiful soup module


### Find and Find all


```python
## Find is method of beautiful soup that only fetches single heading or single html command keywords
first_header = soup.find('h2')

print('Fetching only first header from webpage {}'.format(first_header))

print()

## Find_all is method of beautiful soup that fetches list of the html keywords that given in the argument
headers = soup.find_all('h2')

print('Headers for the webpage {}'.format(headers))
```

```python
print(soup.find(['h1','h2']))  ## find method only fetches first instances of html keyword that given in the argument
print()
soup.find(['h2','h1'])
## In this case In webpage h1 is coming first compare to h2 so whatever may different order of arguments but it pulls first html keyword instance
```

```python
soup.find_all(['h1','h2']) ## It pulls bothe arguments by find_all method
```

```python
## You pass the different argument in the find and find_all methods /
"""
With this help of the argument we can able to fetch the data precisely
"""

paragraph = soup.find_all('p')

print(paragraph,sep="\n",end=' ')  ## In this paragraph we fetch all html keyword that given in argument

## If we want to more specify we use attrs argument

para = soup.find_all('p',attrs={'id':'paragraph-id'})
print()
para
```

```python
### we can able to nest find methods to shortern our web page request
body = soup.find('body')

print('This original body statement \n',body)
print()
div = body.find('div')
print('This scrutern from body webpage keyword  \n',div)
print()
header = div.find('h1')

print('This is filtered from div statement \n',header)
```

```python
## We Can able to find the specific string and string methods from webpage to more significant
import re

string_search = soup.find_all('p',string=re.compile('Some'))

print('String contains specified sub string to fetch from webpage \n',string_search)
print()
headers = soup.find_all('h2',string=re.compile('Header'))
print("Find string  header with captial \n",headers)
print()
header = soup.find_all('h2',string=re.compile('(H|h)eader'))
print('String header with capital and without capital \n',header)
```

```python
## Selecting all paragraph tags in our website

context = soup.select('p')

print('selecting all paragraphs with out any css selector filter \n',context)
print()
content = soup.select(' div p')

print('Selecting specfic paragraph which follows after first key words that given \n',content)

print()

paragraph = soup.select('h2~p')
print('Select specifc paragraph which has same level as first key words that given \n',paragraph)

bold = soup.select('p#paragraph-id b')
print('Selecting the bold letter tag to select that we use # --> mark to select specific paragraph which has id that contains bold value \n',bold)
```

```python
table = webpage.select('table.hockey-stats')[0]
print(table.prettify())
```

```python
import pandas as pd

table = webpage.select('table.hockey-stats')[0]  ### It Returns single element of list that's why we call first and only index of that list
"""
## In here converting scaped data into Data frame
1. Firstly Dataframe needs a df columns or
2. In hockey stats sections there is another table withs its compoenets and that components called as its df columns
3. In order to extract, Using find and find_all methods in soup extracting information from the table
4. Now extracting table rows from that table
"""
#print(table.prettify())

## Extracting the table's columns

columns = table.find('thead').find_all('th')

columns_names = [column.get_text() for column in columns]

print('This is table columns for df but there is repeatation on that table deals this issue out later \n',columns_names)

## Extracting the table's rows

rows = table.find('tbody').find_all('tr')

scrape = []
for i in rows:
  tds = i.find_all('td')
  row = [ tr.get_text().strip()  for tr in tds]
  scrape.append(row)
print()
print(scrape[0])
print()
print('The Scrapped Data Frame \n')
df = pd.DataFrame(scrape,columns=columns_names)

df
```
